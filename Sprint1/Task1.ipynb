### Структура и Объявление Класса в C#

В C#, класс объявляется с использованием ключевого слова `class`, за которым следует имя класса. Класс может содержать поля (данные), свойства (способ доступа к данным), методы (функциональность) и события. 

#### Свойства

Свойства в C# используются для инкапсуляции данных. Они предоставляют гибкий механизм для чтения, записи или вычисления значений приватных полей. Свойства могут иметь два акцессора: `get` (геттер) для чтения значения и `set` (сеттер) для его изменения.

#### Геттеры и Сеттеры

Геттеры и сеттеры являются частью свойств класса в C#. Геттер (`get`) используется для возврата значения свойства, а сеттер (`set`) — для установки нового значения. Сеттер может включать логику валидации или другие операции перед изменением значения поля.

### Пример Классов Vehicle и Car

Давайте рассмотрим примеры классов `Vehicle` и `Car`, чтобы проиллюстрировать использование свойств, геттеров и сеттеров, а также разницу между статическими и обычными атрибутами и методами.

#### Класс Vehicle

```csharp
public class Vehicle
{
    private string _color;
    private int _speed;

    // Свойство Color с геттером и сеттером
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Свойство Speed с геттером и сеттером
    public int Speed
    {
        get { return _speed; }
        set
        {
            if (value >= 0)
                _speed = value;
            else
                throw new ArgumentOutOfRangeException("Скорость не может быть отрицательной!");
        }
    }

    // Статический атрибут
    public static int TotalVehicles { get; private set; } = 0;

    // Конструктор
    public Vehicle()
    {
        TotalVehicles++;
    }

    // Обычный метод
    public void Accelerate(int delta)
    {
        Speed += delta;
    }

    // Статический метод
    public static void DisplayTotalVehicles()
    {
        Console.WriteLine($"Общее количество транспортных средств: {TotalVehicles}");
    }
}
```

#### Класс Car

```csharp
public class Car : Vehicle
{
    private int _fuelLevel;

    // Свойство FuelLevel с геттером и сеттером
    public int FuelLevel
    {
        get { return _fuelLevel; }
        set { _fuelLevel = value; }
    }

    // Конструктор
    public Car() : base()
    {
        FuelLevel = 100; // Задаем начальный уровень топлива
    }

    // Обычный метод
    public void Refuel(int amount)
    {
        FuelLevel += amount;
    }
}
```

### Разница между Статическими и Обычными Атрибутами и Методами Класса

- **Статические атрибуты и методы** принадлежат самому классу, а не его экземплярам. Они могут быть вызваны без создания объекта класса. В примере выше `TotalVehicles` является статическим атрибутом, а `DisplayTotalVehicles()` — статическим методом класса `Vehicle`.
- **Обычные атрибуты и методы** принадлежат экземплярам класса и могут иметь разные значения для разных объектов. В примере выше `_color`, `_speed` и `_fuelLevel` являются обычными атрибутами, а методы `Accelerate()` и `Refuel()` — обычными методами.

````
public class Vehicle
{
    private string _color;
    private int _speed;

    // Свойство Color с геттером и сеттером
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Свойство Speed с геттером и сеттером
    public int Speed
    {
        get { return _speed; }
        set
        {
            if (value >= 0)
                _speed = value;
            else
                throw new ArgumentOutOfRangeException("Скорость не может быть отрицательной!");
        }
    }

    // Статический атрибут
    public static int TotalVehicles { get; private set; } = 0;

    // Конструктор
    public Vehicle()
    {
        TotalVehicles++;
    }

    // Обычный метод
    public void Accelerate(int delta)
    {
        Speed += delta;
    }

    // Статический метод
    public static void DisplayTotalVehicles()
    {
        Console.WriteLine($"Общее количество транспортных средств: {TotalVehicles}");
    }
}

public class Car : Vehicle
{
    private int _fuelLevel;

    // Свойство FuelLevel с геттером и сеттером
    public int FuelLevel
    {
        get { return _fuelLevel; }
        set { _fuelLevel = value; }
    }

    // Конструктор
    public Car() : base()
    {
        FuelLevel = 100; // Задаем начальный уровень топлива
    }

    // Обычный метод
    public void Refuel(int amount)
    {
        FuelLevel += amount;
    }
}
Vehicle.DisplayTotalVehicles();
Vehicle MyVehicle = new Vehicle();
Vehicle.DisplayTotalVehicles();
Console.WriteLine($"У MyVehicle цвет " + MyVehicle.Color);
MyVehicle.Color = "Зелёный";
Console.WriteLine($"У MyVehicle цвет " + MyVehicle.Color);
Console.WriteLine("У MyVehicle скорость " + MyVehicle.Speed);
MyVehicle.Accelerate(17);
Console.WriteLine("У MyVehicle скорость " + MyVehicle.Speed);
MyVehicle.Speed = 50;
Console.WriteLine("У MyVehicle скорость " + MyVehicle.Speed);

Car MyCar = new Car();
Vehicle.DisplayTotalVehicles();
Console.WriteLine($"У MyCar цвет " + MyCar.Color);
MyCar.Color = "Синий";
Console.WriteLine($"У MyCar цвет " + MyCar.Color);
Console.WriteLine("У MyCar скорость " + MyCar.Speed);
MyCar.Accelerate(81);
Console.WriteLine("У MyCar скорость " + MyCar.Speed);
MyCar.Speed = 120;
Console.WriteLine("У MyCar скорость " + MyCar.Speed);
Console.WriteLine("У MyCar " + MyCar.FuelLevel + " топлива");
MyCar.Refuel(19);
Console.WriteLine("У MyCar " + MyCar.FuelLevel + " топлива");
````

````
public class Animal
{
    public static int TotalAnimals { get; set; } = 0;

    private int size;
    private int weight;
    public int Age {get; private set;}

    public Animal(int Size, int Weight, int Age)
    {
        this.Size = Size;
        this.Weight = Weight;
        this.Age = Age;
        TotalAnimals++;
    }

    public int Size
    {
        get { return size; }
        private set { size = value; }
    }

    public int Weight
    {
        get { return weight; }
        set
        {
            if (value > -1)
                weight = value;
            else
                throw new ArgumentOutOfRangeException("Масса не может быть отрицательной");
        }
    }

    public void GrowUp(int Age)
    {
        if (Age > -1)
        {
            this.Age += Age;
        }
        else
        {
            throw new ArgumentOutOfRangeException("Нельзя постареть на отрицательное количество лет");
        }
    }

    public static void DisplayTotalAnimals()
    {
        Console.WriteLine($"Общее количество зверей: {TotalAnimals}");
    }
}

public class Rabbit : Animal
{
    private static int TotalRabbits {get; set;} = 0;
    public int EarLength;
    private int carrotNumber;

    public Rabbit(int Size, int Weight, int Age, int EarLength) : base(Size, Weight, Age)
    {
        this.EarLength = EarLength;
        CarrotNumber = 0;
        TotalRabbits++;
    }

    public int CarrotNumber
    {
        get { return carrotNumber; }
        set { carrotNumber = value; }
    }

    public void DigAHome(int Depth)
    {
        Console.WriteLine("Вырыта нора глубиной " + Depth + " метров");
    }

    public static void DisplayTotalRabbits()
    {
        Console.WriteLine("Общее количество кроликов: " + TotalRabbits);
    }
}

public class Fox : Animal
{
    private static int TotalFoxes {get; set;} = 0;
    public int EyeQuality;
    public string Location { get; private set; }

    public Fox(int Size, int Weight, int Age, int EyeQuality, string Location) : base(Size, Weight, Age)
    {
        this.EyeQuality = EyeQuality;
        this.Location = Location;
        TotalFoxes++;
    }

    public static void DisplayTotalFoxes()
    {
        Console.WriteLine("Общее количество лисов: " + TotalFoxes);
    }

    public void Hunt(int PrayId)
    {
        Console.WriteLine("Лиса охотится на добычу с id=" + PrayId);
    }

    public void Travel(string Location)
    {
        this.Location = Location;
    }
}

Animal.DisplayTotalAnimals();
Rabbit.DisplayTotalRabbits();
Rabbit MyRabbit = new Rabbit(5, 10, 2, 10);
Animal.DisplayTotalAnimals();
Rabbit.DisplayTotalRabbits();
Console.WriteLine("У MyRabbit размер " + MyRabbit.Size);
Console.WriteLine("У MyRabbit масса " + MyRabbit.Weight);
MyRabbit.Weight = 12;
Console.WriteLine("У MyRabbit масса " + MyRabbit.Weight);
Console.WriteLine("У MyRabbit возраст " + MyRabbit.Age);
MyRabbit.GrowUp(1);
Console.WriteLine("У MyRabbit возраст " + MyRabbit.Age);
Console.WriteLine("У MyRabbit уши длиной " + MyRabbit.EarLength);
Console.WriteLine("У MyRabbit размер " + MyRabbit.Size);
Console.WriteLine("У MyRabbit " + MyRabbit.CarrotNumber + " моркови");
MyRabbit.CarrotNumber = 2;
Console.WriteLine("У MyRabbit " + MyRabbit.CarrotNumber + " моркови");
MyRabbit.DigAHome(3);

Animal.DisplayTotalAnimals();
Rabbit.DisplayTotalRabbits();
Fox.DisplayTotalFoxes();
Fox MyFox = new Fox(5, 10, 3, 7, "Лес");
Animal.DisplayTotalAnimals();
Rabbit.DisplayTotalRabbits();
Fox.DisplayTotalFoxes();
Console.WriteLine("У MyFox размер " + MyFox.Size);
Console.WriteLine("У MyFox масса " + MyFox.Weight);
MyFox.Weight = 12;
Console.WriteLine("У MyFox масса " + MyFox.Weight);
Console.WriteLine("У MyFox возраст " + MyFox.Age);
MyFox.GrowUp(1);
Console.WriteLine("У MyFox возраст " + MyFox.Age);
Console.WriteLine("У MyFox зрение " + MyFox.EyeQuality);
MyFox.GrowUp(1);
Console.WriteLine("MyFox находится в " + MyFox.Location);
MyFox.Travel("Дубовая роща");
Console.WriteLine("MyFox находится в " + MyFox.Location);
MyFox.Hunt(41);
````
