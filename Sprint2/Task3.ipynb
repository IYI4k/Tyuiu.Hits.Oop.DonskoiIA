<h2 style="color:DodgerBlue">Описание проекта:</h2>

----
Создать базовый класс Item в C#, который будет представлять информацию о  
товарах, которые могут быть заказаны или возвращены. На основе этого класса  
разработать 2-3 производных класса, демонстрирующих принципы наследования и  
полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и  
методы, а также переопределены некоторые методы базового класса для  
демонстрации полиморфизма. 
Требования к базовому классу Item: 
• Атрибуты: ID товара (ItemId), Название (Name), Цена (Price). 
• Методы: 
o 
o GetDetails(): метод для получения детальной информации о товаре. 
o CalculateDiscount(): метод для расчета скидки на товар. 
o ApplyDiscount(decimal discount): метод для применения скидки к цене  
товара. 
Требования к производным классам: 
1. ЕдиничныйТовар (SingleItem): Должен содержать дополнительные атрибуты,  
такие как Единица измерения (UnitMeasure). Метод GetDetails() должен быть  
переопределен для добавления информации о единице измерения товара. 
2. ПакетныйТовар (PackageItem): Должен содержать дополнительные  
атрибуты, такие как Количество единиц в пакете (QuantityPerPackage).  
Метод CalculateDiscount() должен быть переопределен для учета количества  
единиц в пакете при расчете скидки. 
3. СпециальныйТовар (SpecialItem) (если требуется третий класс): Должен  
содержать дополнительные атрибуты, такие как Дата истечения скидки  
(DiscountExpirationDate). Метод ApplyDiscount() должен быть переопределен  
для добавления информации о сроке действия скидки.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
class item
{
    private static int TotalItems = 0;
    public int ItemId { get; private set;}
    public string Name { get; set; }
    public int Price { get; protected set;}
    public string Location { get; set; }
    public string AppearDate { get; set; }

    public item(int price) {
        Price = price;
        ItemId = TotalItems++;
    }

    public virtual string GetDetails()
    {
        return $"ItemId={ItemId};Name={Name};Price={Price};Location={Location};AppearDate={AppearDate}";
    }

    public virtual decimal CalculateDiscount(decimal discount)
    {
        return Price * (1 - discount);
    }

    public void ApplyDiscount(decimal discount)
    {
        Price = (int)(Price * (1 - discount));
    }

    public int GetPrice()
    {
        return Price;
    }

    public void SetPrice(int price)
    {
        Price = price;
    }
}

public interface IPrint
{
    public void Print();
}

class SingleItem : item, IPrint
{
    public string UnitMeasure { get; private set; }
    public int count { get; private set; }

    public SingleItem(int price, int count) : base(price)
    {
        this.count = count;
    }

    public void SetDetails(string unitMeasure)
    {
        UnitMeasure = unitMeasure;
    }

    public void SetCount(int Count)
    {
        count = Count;
    }

    public string GetUnitMeasure() {
        return UnitMeasure;
    }
    
    public void Print()
    {
        Console.WriteLine("Идёт печать информации о единичном товаре");
    }
    
    public string GetInfo()
    {
        return $"ItemId={ItemId};Name={Name};Price={Price};Count={count};UnitMeasure={UnitMeasure}";
    }
}

public interface ICalculateAxB
{
    public decimal CalculateAxB(decimal a, decimal b);
}

class MathCalculateAxB : ICalculateAxB
{
    public decimal CalculateAxB(decimal a, decimal b)
    {
    return a * b;
    }
}

class PackageItem : item, IPrint
{
    public int QuanityPerPackage { get; set; }
    ICalculateAxB _calculateAxB;
    public int Popularity { get; set; }
    public int ValueInUnitMeasure { get; set; }

    public PackageItem(ICalculateAxB calculateAxB, int price, int quanityPerPackage) : base(price)
    {
        _calculateAxB = calculateAxB;
        QuanityPerPackage = quanityPerPackage;
    }

    public override decimal CalculateDiscount(decimal discount)
    {
        return base.CalculateDiscount(discount) * QuanityPerPackage;
    }

    public decimal GetTotalPrice()
    {
        return _calculateAxB.CalculateAxB((decimal)GetPrice(), (decimal)QuanityPerPackage);
    }
    
    public void Print()
    {
        Console.WriteLine("Идёт печать информации о пакетном товаре");
    }

    public string GetInfo()
    {
        return $"ItemId={ItemId};Name={Name};Price={Price};QuanityPerPackege={QuanityPerPackage};Popularity={Popularity};ValueInUnitMeasure={ValueInUnitMeasure}";
    }
}

class Pen : SingleItem
{
    public string LineThickness { get; set; }
    
    public Pen(int price, int count, string lineThickness) : base(price, count)
    {
        LineThickness = lineThickness;
    }
}

item it = new item(900);
Console.WriteLine(it.GetDetails());
Console.WriteLine(it.CalculateDiscount((decimal)0.1));
it.ApplyDiscount((decimal)0.1);
it.Location="Тюмень";
it.AppearDate="02.10.2025";
Console.WriteLine(it.GetDetails());
Console.WriteLine(it.GetPrice());
it.SetPrice(1000);
Console.WriteLine(it.GetPrice());

SingleItem singleItem = new SingleItem(100, 10000);
singleItem.SetDetails("Кг");
singleItem.Location="Томск";
singleItem.AppearDate="04.10.2025";
Console.WriteLine(singleItem.GetUnitMeasure());
singleItem.SetCount(5);
Console.WriteLine(singleItem.count);
singleItem.Print();
Console.WriteLine(singleItem.GetInfo());

MathCalculateAxB calculateAxB = new MathCalculateAxB();
PackageItem packageItem = new PackageItem(calculateAxB, 60, 7);
packageItem.ValueInUnitMeasure = 2;
Console.WriteLine(packageItem.CalculateDiscount((decimal)0.33));
Console.WriteLine(packageItem.GetTotalPrice());
packageItem.Location="Тобольск";
packageItem.AppearDate="01.10.2025";
packageItem.Print();
Console.WriteLine(packageItem.GetInfo());

Pen pen = new Pen(58, 100, "0.7");
pen.Print();
pen.GetInfo();
```
